# Exploring eBay Car Sales Data

In this Python project, we will work with a dataset of used cars from eBay Kleinanzeigne, a classfields section of the German eBay website.

The data set was originally scraped and uploaded to Kaggle by user orgesleka. The original dataset isn't available on Kaggle anymore, but you can find it [here](https://data.world/data-society/used-cars-data)

The data dictionary provided with data is as follows:

- dateCrawled - When this ad was first crawled. All field-values are taken from this date.
- name - Name of the car.
- seller - Whether the seller is private or a dealer.
- offerType - The type of listing
- price - The price on the ad to sell the car.
- abtest - Whether the listing is included in an A/B test.
- vehicleType - The vehicle Type.
- yearOfRegistration - The year in which the car was first registered.
- gearbox - The transmission type.
- powerPS - The power of the car in PS.
- model - The car model name.
- kilometer - How many kilometers the car has driven.
- monthOfRegistration - The month in which the car was first registered.
- fuelType - What type of fuel the car uses.
- brand - The brand of the car.
- notRepairedDamage - If the car has a damage which is not yet repaired.
- dateCreated - The date on which the eBay listing was created.
- nrOfPictures - The number of pictures in the ad.
- postalCode - The postal code for the location of the vehicle.
- lastSeenOnline - When the crawler saw this ad last online.

Read autos.csv CSV file into pandas, and assign it to the variable name autos

````python
autos = pd.read_csv("autos.csv", encoding="Latin-1")
````

Now, use the DataFrame.info() and DataFrame.head() methods to print information about the ```autos``` dataframe, as well as the first few rows. 

````python
autos.info()
autos.head()
````

The column names use camelcase instead of Python's preferred snakecase. In the codes below, we are going to convert camelcase into snakecase. 


````python
autos.rename(columns={"yearOfRegistration":"registration_year", "monthOfRegistration":"registration_month", "notRepairedDamage":'unrepaired_damage', "dateCreated":"ad_created"}, inplace = True)


def change_case(str):
    res = [str[0].lower()]
    for c in str[1:]:
        if c in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
            res.append('_')
            res.append(c.lower())
        else:
            res.append(c)
     
    return ''.join(res)

new_columns = []
columns = autos.columns

for column in columns: 
    new_column = change_case(column)
    new_columns.append(new_column)
    
autos.columns = new_columns    
````

In the below codes, we are going to use DataFrame.describe() to look at descriptive statistics for all columns. 

````python
autos.describe(include='all')
````

We can note that the ```price``` and ```odometer``` columns are numeric values stored as text. 

We will first convert the value in the "price" column into integer values. 

````python
new_prices = []

for price in autos["price"]:
    price = price.replace("$", "")
    price = price.replace(",","")
    price = int(price)
    new_prices.append(price)

autos["price"] = new_prices
    
    
````

Now we will execute the same thing for the ```odometer``` column.

````python
new_odometers = []

for odometer in autos["odometer"]:
    odometer = odometer.replace("km", "")
    odometer = odometer.replace(",","")
    odometer = int(odometer)
    new_odometers.append(odometer)

autos["odometer"] = new_odometers
autos.rename(columns={"odometer":"odometer_km"}, inplace=True)
````

Now that we have cleaned up the ```price``` and ```odometer_km``` column, let's further explore the data. We are going to specifically look for data that doesn't look right. 

````python


print(autos["price"].unique())
print(autos["price"].describe())
print(autos["price"].value_counts())
````

Based on the analysis above, we can find outliers for the ```price``` column. We will remove the outlier below: 

````python
autos[(autos["price"] <= 2500)]
````

We will repeate the same process for the ```odometer_km``` column.

````python
print(autos["odometer_km"].unique())
print(autos["odometer_km"].describe())
print(autos["odometer_km"].value_counts())
````

Based on the analysis above, we can conclude that the ```odometer_km``` column does not have an outlier that we have to remove.
